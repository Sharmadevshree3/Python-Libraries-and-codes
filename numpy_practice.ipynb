**MATRIX**

```python
Matrix=np.array([[1,1,2],[2,4,5],[3,6,2]])
Matrix
```

```python
print(Matrix.transpose())     #transpose
```

```python
np.linalg.inv(Matrix)    #inverse
```

```python
np.linalg.det(Matrix)      #determinant
```

```python
np.linalg.eig(Matrix)
# Eigen values
```

```python
np.trace(Matrix)
```

```python
Mat=np.array([[2,3,2],[8,6,5],[2,1,7]])
```

```python
np.dot(Matrix,Mat)
# Dot product
```

```python
np.matmul(Matrix,Mat)
# Matrix multiplication
```

```python
np.linalg.solve(Matrix,Mat)
# Linear equations
```
